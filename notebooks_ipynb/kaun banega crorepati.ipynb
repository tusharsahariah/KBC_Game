```python

for i in range(df.shape[0]):
    question=questions[i][0]
    a=questions[i][1][0]
    b=questions[i][1][1]
    c=questions[i][1][2]
    d=questions[i][1][3]
    correct=questions[i][2]
    lifeline=['L1','L2','L3','L4']


    q=df.loc[i,'Q.No']
    p=df.loc[i,'Prize Money']

    print(f'Q{q}. for {p}:💵')
    print(f'{question}\nA.{a:10}     B.{b:10}\nC.{c:10}     D.{d:10}\n ')
    command=input("Enter Your Answer:").upper()
    if command==correct:
        print("Bilkul sahi jawab! 🎉")
        print(f"✅Ab aap jeet chuke hain {p}!")
        print('🗣️NEXT QUESTION AAPKE COMPUTER SCREEN PAR YEH RAHA \n')
        continue
        
    elif command=="Q":
        
        if i==0:
            print(f"🤝Computer ji, khiladi ne quit karne ka faisla kiya hai.\nKEEP IT UP👍")
        else:
            print(f"🤝Computer ji, khiladi ne quit karne ka faisla kiya hai.\nAap le kar ja rahe hain {p}.")
        break
    else:
        if i==0:
            print(f"❌Afsos, yeh galat jawab hai.\nAap le kar ja rahe hain kuch nahi.NOOOOOOB!😂\nGAME ENDS!!🥁")
        else:
            print(f"❌Afsos, yeh galat jawab hai.\nAap le kar ja rahe hain {p}\nGAME ENDS!!🥁")
        break
else:
    print('Mubarak ho! Aap ban gaye hain Crorepati!!!\n🎶 [Victory music] 🎶')
```
